```python
import pandas as pd
df = pd.read_csv('path_to_your_file.csv')

X = df['X']
delta0_4_s = df['delta0_4_s']
delta0_4 = df['delta0_4']
delta0_4_real = df['delta0_4_real']
omega0_4_s = df['omega0_4_s']
omega0_4 = df['omega0_4']
omega_0_4_real = df['omega_0_4_real']
```

```python
import matplotlib.pyplot as plt

# Plot for delta
plt.figure()
plt.scatter(X, delta0_4_s, label="SFQ $\delta$")
plt.plot(X, delta0_4, 'r--', label="PFQ $\delta$")
plt.plot(X, delta0_4_real, color='g', label="Actual $\delta$")
plt.xlabel('$t$', fontsize=16)
plt.ylabel('$\delta$(t)', fontsize=16)
plt.legend(fontsize=14)
plt.show()

# Plot for omega
plt.figure()
plt.scatter(X, omega0_4_s, label="SFQ $\omega$")
plt.plot(X, omega0_4, 'r--', label="PFQ $\omega$")
plt.plot(X, omega_0_4_real, color='g', label="Actual $\omega$")
plt.xlabel('$t$', fontsize=16)
plt.ylabel('$\omega(t)$', fontsize=16)
plt.legend(fontsize=14)
plt.show()
```

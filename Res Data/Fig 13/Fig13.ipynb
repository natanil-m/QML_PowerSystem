```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data from CSV
data = pd.read_csv('data.csv')

# Extract data
X = data['X']
delta0_8 = data['delta0_8']
X_real = data['X_real']
delta0_8_real = data['delta0_8_real']
omega0_8 = data['omega0_8']
omega_0_8_real = data['omega_0_8_real']


#rescale
delta0_8 = np.array(delta0_8)*10
delta0_8_real = np.array(delta0_8_real)*10
omega0_8 = np.array(omega0_8)*10
omega_0_8_real = np.array(omega_0_8_real)*10




# Set font size
font_size = 14

# Reproduce the delta plot
plt.figure()
plt.plot(X, delta0_8, 'r--', label="Approximation $\delta$")
plt.plot(X_real, delta0_8_real, color='g', label="Actual $\delta$")
plt.xlabel('$t$', fontsize=16)
plt.ylabel('$\delta$(t)', fontsize=16)
plt.legend(fontsize=font_size)

# Reproduce the omega plot
plt.figure()
plt.plot(X, omega0_8, 'r--', label="Quantum $\omega$")
plt.plot(X_real, omega_0_8_real, color='g', label="Actual $\omega$")
plt.xlabel('$t$', fontsize=16)
plt.ylabel('$\omega(t)$', fontsize=16)
plt.legend(fontsize=font_size)
plt.show()
```

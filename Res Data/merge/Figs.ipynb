```python
import pandas as pd
df = pd.read_csv('data_f11.csv')

X = df['X']
delta0_4_s = df['delta0_4_s']
delta0_4 = df['delta0_4']
delta0_4_real = df['delta0_4_real']
omega0_4_s = df['omega0_4_s']
omega0_4 = df['omega0_4']
omega_0_4_real = df['omega_0_4_real']
```

```python
import matplotlib.pyplot as plt

# Plot for delta
plt.figure()
plt.scatter(X, delta0_4_s, label="SFQ $\delta$")
plt.plot(X, delta0_4, 'r--', label="PFQ $\delta$")
plt.plot(X, delta0_4_real, color='g', label="Actual $\delta$")
plt.xlabel('$t$', fontsize=16)
plt.ylabel('$\delta$(t)', fontsize=16)
plt.legend(fontsize=14)
plt.show()

# Plot for omega
plt.figure()
plt.scatter(X, omega0_4_s, label="SFQ $\omega$")
plt.plot(X, omega0_4, 'r--', label="PFQ $\omega$")
plt.plot(X, omega_0_4_real, color='g', label="Actual $\omega$")
plt.xlabel('$t$', fontsize=16)
plt.ylabel('$\omega(t)$', fontsize=16)
plt.legend(fontsize=14)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
import numpy as np

# Read the data for actual values
actual_data = pd.read_csv('actual_data_3pts.csv')
t_data_actual = actual_data['t_data_actual']
omega_target_values = actual_data['omega_target_values']

# Read the data for predicted values
predicted_data = pd.read_csv('predicted_data_3pts.csv')
t_data = predicted_data['t_data']
delta_pred = predicted_data['delta_pred']
omega_pred = predicted_data['omega_pred']

# # Interpolate predicted omega values to match the timeline of t_data_actual
interp_function = interp1d(t_data, omega_pred, kind='linear', fill_value="extrapolate")
omega_pred_interpolated = interp_function(t_data_actual)

#rescale
omega_target_values = np.array(omega_target_values)*10
omega_pred_interpolated = np.array(omega_pred_interpolated)*10

# Set a larger font size
font_size = 13

# Create a figure and plot
plt.figure(figsize=(5, 5))

# Plotting Actual Omega Values
plt.plot(t_data_actual, omega_target_values, color='green', label='Classical Solution (Actual)', alpha=0.6)

# Plotting Interpolated Predicted Omega Values
plt.plot(t_data_actual, omega_pred_interpolated, 'r--', label='Quantum Solution (Predicted)')

# Setting labels and legend
plt.xlabel('$t$', fontsize=font_size)
plt.ylabel('$\omega(t)$', fontsize=font_size)
plt.legend(fontsize=font_size)
plt.title('Comparison of Actual and Predicted Omega Values')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data from CSV
data = pd.read_csv('data_13.csv')

# Extract data
X = data['X']
delta0_8 = data['delta0_8']
X_real = data['X_real']
delta0_8_real = data['delta0_8_real']
omega0_8 = data['omega0_8']
omega_0_8_real = data['omega_0_8_real']


#rescale
delta0_8 = np.array(delta0_8)*10
delta0_8_real = np.array(delta0_8_real)*10
omega0_8 = np.array(omega0_8)*10
omega_0_8_real = np.array(omega_0_8_real)*10




# Set font size
font_size = 14

# Reproduce the delta plot
plt.figure()
plt.plot(X, delta0_8, 'r--', label="Approximation $\delta$")
plt.plot(X_real, delta0_8_real, color='g', label="Actual $\delta$")
plt.xlabel('$t$', fontsize=16)
plt.ylabel('$\delta$(t)', fontsize=16)
plt.legend(fontsize=font_size)

# Reproduce the omega plot
plt.figure()
plt.plot(X, omega0_8, 'r--', label="Quantum $\omega$")
plt.plot(X_real, omega_0_8_real, color='g', label="Actual $\omega$")
plt.xlabel('$t$', fontsize=16)
plt.ylabel('$\omega(t)$', fontsize=16)
plt.legend(fontsize=font_size)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data from CSV
data = pd.read_csv('data_14.csv')

# Extract data
t_data = data['t_data']
del1_target_values = data['del1_target_values']
del1 = data['del1']
omega1_target_values = data['omega1_target_values']
omega1 = data['omega1']
del2_target_values = data['del2_target_values']
del2 = data['del2']
omega2_target_values = data['omega2_target_values']
omega2 = data['omega2']
del3_target_values = data['del3_target_values']
del3 = data['del3']
omega3_target_values = data['omega3_target_values']
omega3 = data['omega3']

# Create a 3x2 grid of subplots with padding and border
fig, axs = plt.subplots(3, 2, figsize=(20, 14), gridspec_kw={'hspace': 0.04, 'wspace': 0.06})

# Adjust layout for better spacing
# plt.tight_layout()

# Add padding and border to each subplot
for ax in axs.flatten():
    ax.set_position([ax.get_position().x0, ax.get_position().y0, ax.get_position().width, ax.get_position().height])
    ax.set_position([ax.get_position().x0 + 0.05, ax.get_position().y0 + 0.05, ax.get_position().width - 0.05, ax.get_position().height - 0.05])


#Initial Condition Setup:
del1[0]=del1_target_values[0]    
omega1[0]=omega1_target_values[0]   
del2[0]=del2_target_values[0]    
omega2[0]=omega2_target_values[0]   
del3[0]=del3_target_values[0]    
omega3[0]=omega3_target_values[0]   

fontsize=17

# Plot for delta1_omega1
axs[0, 0].plot(t_data, del1_target_values, 'g',label='Actual $\delta_1$', alpha=1)
axs[0, 0].plot(t_data, del1, 'r--',label='Approximation $\delta_1$')
axs[0, 0].set_xlabel('$t$', fontsize=fontsize)
axs[0, 0].set_ylabel('$\delta_1$', fontsize=fontsize)
axs[0, 0].legend(fontsize=fontsize)
axs[0, 0].grid(True)

axs[0, 1].plot(t_data, omega1_target_values, 'g',label='Actual $\omega_1$', alpha=1)
axs[0, 1].plot(t_data, omega1,'r--',label='Approximation $\omega_1$')
axs[0, 1].set_xlabel('$t$', fontsize=fontsize)
axs[0, 1].set_ylabel('$\omega_1$', fontsize=fontsize)
axs[0, 1].legend(fontsize=fontsize)
axs[0, 1].grid(True)
axs[0, 1].set_ylim([min(omega1_target_values)-0.00001, max(omega1_target_values)+0.00001])

# Plot for delta2_omega2
axs[1, 0].plot(t_data, del2_target_values, 'g',label='Actual $\delta_2$', alpha=1)
axs[1, 0].plot(t_data, del2, 'r--',label='Approximation $\delta_2$')
axs[1, 0].set_xlabel('$t$', fontsize=fontsize)
axs[1, 0].set_ylabel('$\delta_2$', fontsize=fontsize)
axs[1, 0].legend(fontsize=fontsize)
axs[1, 0].grid(True)

axs[1, 1].plot(t_data, omega2_target_values, 'g',label='Actual $\omega_2$', alpha=1)
axs[1, 1].plot(t_data, omega2,'r--', linewidth=1,label='Approximation $\omega_2$')
axs[1, 1].set_xlabel('$t$', fontsize=fontsize)
axs[1, 1].set_ylabel('$\omega_2$', fontsize=fontsize)
axs[1, 1].legend(fontsize=fontsize)
axs[1, 1].grid(True)
axs[1, 1].set_ylim([min(omega2_target_values)-0.0001, max(omega2_target_values)+0.0001])

# Plot for delta3_omega3
axs[2, 0].plot(t_data, del3_target_values, 'g',label='Actual $\delta_3$', alpha=1)
axs[2, 0].plot(t_data, del3,'r--',label='Approximation $\delta_3$')
axs[2, 0].set_xlabel('$t$', fontsize=fontsize)
axs[2, 0].set_ylabel('$\delta_3$', fontsize=fontsize)
axs[2, 0].legend(fontsize=fontsize)
axs[2, 0].grid(True)

axs[2, 1].plot(t_data, omega3_target_values, 'g',label='Actual $\omega_3$', alpha=1)
axs[2, 1].plot(t_data, omega3,'r--',label='Approximation $\omega_3$')
axs[2, 1].set_xlabel('$t$', fontsize=fontsize)
axs[2, 1].set_ylabel('$\omega_3$', fontsize=fontsize)
axs[2, 1].legend(fontsize=fontsize)
axs[2, 1].grid(True)
axs[2, 1].set_ylim([min(omega3_target_values)-0.0002, max(omega3_target_values)+0.0002])

# Show the plot
plt.show()
```
